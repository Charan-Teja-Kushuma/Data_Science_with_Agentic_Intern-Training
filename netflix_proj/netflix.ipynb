```python
#declaring or casting column 
df['Release_Date'] = pd.to_datetime(df['Release_Date'])

#confirming changes
print(df['Release_Date'].dtypes)
                                    
```

```python
#droping uncessary columns in dataset (overview, Original Lang, poster url
cols = ['Overview', 'Original_Language', 'Poster_Url']
```

```python
df.drop(cols, axis = 1, inplace = True)
df.columns
```

```python
df.head()
```

```python
#Checking categorizing Rating 
#Not Popular
#Below Average
#Average
#popular

def catigorize_col(df, col, labels):
    edges = [
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].describe()['max']
    ]
    df[col] = pd.cut(df[col], edges, labels = labels, duplicates='drop')
    return df
```

```python
#define labels for edges 
labels = ['not_popular', 'below_avg', 'average', 'popular']

#catigorize columns based lables and edges passign on function
catigorize_col(df, 'Vote_Average', labels)

#confirming changes
df['Vote_Average'].unique()
```

```python
#exploring specific coluymns
df['Vote_Average'].value_counts()
```

```python
#dropping NAN values
df.dropna(inplace = True)

#confirming nan values
df.isna().sum()
```

```python
df.head()
```

```python
#split the string into list
df['Genre'] = df['Genre'].str.split(',')
```

```python
df.head()
```

```python
#explode list
df = df.explode('Genre').reset_index(drop=True)
```

```python
df.head()
```

```python
# genre column into category 
df['Genre'] = df['Genre'].astype('category')
```

```python
#confirming chages
df['Genre'].dtypes
```

```python
df.info()
```

```python
#data visualization
#settings up seaborn configutation

sns.set_style('whitegrid')
```

```python
#showing stats on genre column
df['Genre'].describe()
```

```python
#visualizing genre column
sns.catplot(y='Genre', data=df, kind= 'count',
            order = df['Genre'].value_counts().index,
            color = '#4285f5')
plt.title('Genre column distribution')
plt.show()
```

```python
#checking highest votes in genre
sns.catplot(y='Vote_Average', data= df, kind = 'count',
             order = df['Vote_Average'].value_counts().index,
             color = '#4285f5')
plt.title('Vote Distribution')
plt.show()
```
